```python
from mnist_loader import load_mnist
from image_ops import mnist_image_display, normalize_image_array
from models import SingleLayerNet, TwoLayerNet, ThreeLayerNet

print("All necessary packages have been imported!")
```

```python
x_train, x_test, t_train, t_test = load_mnist()

print("Check the size of the training set: ",len(x_train))
print("Check the size of the test set: ",len(x_test))
print("Check the size of the training label: ",len(t_train))
print("Check the size of the test label: ",len(t_test))
```

```python
mnist_image_display(10, 10, x_train, t_train)

print("The illustration of data of class 0 - 9 can be found below.")
```

```python
x_train = normalize_image_array(x_train)
```

```python
net1 = SingleLayerNet(x_train=x_train, num_w1_node=10)
net1.predict()
net1.calculate_accuracy(t_train=t_train)
```

```python
import numpy as np
random_idx = np.random.choice(x_train.shape[0], 500)
```

```python
net2 = TwoLayerNet(num_in_node=784, num_w1_node=20, num_w2_node=10)
net2.predict(x_train[random_idx])
net2.accuracy(x_train[random_idx], t_train[random_idx])
```

```python
net2.loss(x_train[random_idx], t_train[random_idx])
```

```python
net2.gradient(x_train[random_idx], t_train[random_idx])
```

```python
net3 = ThreeLayerNet(x_train=x_train, num_w1_node=300, num_w2_node=100, num_w3_node=10)
net3.predict()
net3.calculate_accuracy(t_train=t_train)
```
